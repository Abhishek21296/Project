```python
import tensorflow as tf
import numpy as np
import pandas as pd
import tflearn as tl
```

```python
raw_data = pd.read_csv("/content/drive/My Drive/mnist_train.csv")
```

```python
raw_data.head()
```

```python
labels=raw_data.iloc[:,0]
```

```python
raw_data.drop(labels='label',axis=1, inplace=True)
```

```python
x=np.array(raw_data)
```

```python
c=np.eye(10,10)[labels]
```

```python
B=128
epochs=50
iterations=int(x.shape[0]/B)
```

```python
c
```

```python
neurons=[x.shape[1],x.shape[1],x.shape[1],x.shape[1],c.shape[1]]
```

```python
layer_activations=[]
layer0_activations=tl.layers.core.input_data(shape=[None,neurons[0]])
layer_activations.append(layer0_activations)
```

```python
for i in range(len(neurons)-2):
    intermidiate_output=tl.layers.core.fully_connected(incoming=layer_activations[i],n_units=neurons[i+1],activation='sigmoid')
    layer_activations.append(tl.layers.normalization.batch_normalization(incoming=intermidiate_output))
```

```python
layer_activations
```

```python
output_layer=tl.layers.core.fully_connected(incoming=layer_activations[-1],n_units=neurons[-1],activation='softmax')
layer_activations.append(output_layer)
```

```python
loss_layer=tl.layers.estimator.regression(incoming=layer_activations[-1],batch_size=B,n_classes=c.shape[1],optimizer='SGD')
```

```python
nn_model=tl.models.dnn.DNN(network=loss_layer)
```

```python
plchldr=tf.get_collection(tf.GraphKeys.TARGETS)
plchldr0=tf.get_collection(tf.GraphKeys.INPUTS)
nn_model.fit(X_inputs={plchldr0[0]:x},Y_targets={plchldr[0]:c},n_epoch=epochs)
```
