```python
import pandas as pd 
import numpy as np
```

```python
raw_data = pd.read_csv("./data.csv")
```

```python
raw_data.head()
```

```python
X = raw_data.drop(['diagnosis','id',raw_data.columns[32]],axis=1)
```

```python
X
```

```python
X = np.array(X)
```

```python
X = (X-np.mean(X,axis=0))/np.std(X,axis=0)
```

```python
X
```

```python
C = raw_data['diagnosis'].replace(to_replace=['B','M'],value=[0,1])
```

```python
C = np.array(C)
```

```python
def sigmoid(z):
    
    return 1/(1 + np.exp(-z))
```

```python
def layer_activations(last_activations,current_thetas,current_theta0s):
    
    current_activations = sigmoid(np.matmul(last_activations,current_thetas)+current_theta0s)
    
    return current_activations
```

```python
def forward_layer_derivatives(last_activations,current_activations):
    
    return np.matmul(last_activations.T,(current_activations*(1-current_activations)))
```

```python
def forward_layer_derivatives0(current_activations):
    
    del0 = np.sum(current_activations*(1-current_activations),axis=0)
    
    del0 = del0.reshape(1,del0.shape[0])
    
    return del0
```

```python
def reverse_layer_derivatives(last_reverse_derivatives,last_thetas,last_layer_activations):
    
    return ((last_reverse_derivatives.T)*np.matmul(last_thetas,((last_layer_activations)*(1-last_layer_activations)).T)).T
```

```python
def neg_log_loss(C_batch,nn_out):
    
    return -(1/B)*(np.matmul(C_batch.T,np.log(nn_out)) + np.matmul((1-C_batch).T,np.log(1-nn_out)))
```

```python
alpha = 0.0001

i = 0

iterations = []

max_i = 50000

neg_log_loss_history = []

B = 32

neurons = [X.shape[1],X.shape[1],X.shape[1],1]

theta0s_initial = [0,np.random.randn(1,neurons[1]),np.random.randn(1,neurons[2]),np.random.randn(1,neurons[3])]

thetas_initial = [0,np.random.randn(neurons[0],neurons[1]),np.random.randn(neurons[1],neurons[2]),np.random.randn(neurons[2],neurons[3])]
```

```python
while i< max_i:
    
    random_indices = np.random.randint(low=0,high=X.shape[0],size=B)

    l0_activations = X[random_indices]



    l1_activations = layer_activations(l0_activations,thetas_initial[1],theta0s_initial[1])

    l1_forward_dels = forward_layer_derivatives(l0_activations,l1_activations)

    l1_forward_dels0 = forward_layer_derivatives0(l1_activations)
    



    l2_activations = layer_activations(l1_activations,thetas_initial[2],theta0s_initial[2])

    l2_forward_dels = forward_layer_derivatives(l1_activations,l2_activations)

    l2_forward_dels0 = forward_layer_derivatives0(l2_activations)
    



    l3_activations = layer_activations(l2_activations,thetas_initial[3],theta0s_initial[3])

    l3_forward_dels = forward_layer_derivatives(l2_activations,l3_activations)

    l3_forward_dels0 = forward_layer_derivatives0(l3_activations)






    C_batch = C[random_indices]

    C_batch = C_batch.reshape(C_batch.shape[0],1)

    E = l3_activations - C_batch



    l3_reverse_dels = E/(l3_activations*(1-l3_activations))

    l3_reverse_dels = np.sum(l3_reverse_dels)



    l2_reverse_dels = np.sum(E*thetas_initial[3].T,axis=0)

    l2_reverse_dels = l2_reverse_dels.reshape(1,l2_reverse_dels.shape[0])



    l1_reverse_dels = np.sum(reverse_layer_derivatives(l2_reverse_dels,thetas_initial[2],l2_activations),axis=0)

    l1_reverse_dels = l1_reverse_dels.reshape(1,l1_reverse_dels.shape[0])





    overall_l1_dels = (l1_forward_dels*l1_reverse_dels)/B

    overall_l1_dels0 = (l1_forward_dels0*l1_reverse_dels)/B

    overall_l2_dels = (l2_forward_dels*l2_reverse_dels)/B

    overall_l2_dels0 = (l2_forward_dels0*l2_reverse_dels)/B

    overall_l3_dels = (l3_forward_dels*l3_reverse_dels)/B

    overall_l3_dels0 = (l3_forward_dels0*l3_reverse_dels)/B
    
    
    
    
    thetas_final_l1 = thetas_initial[1] - (alpha * overall_l1_dels)
    
    theta0s_final_l1 = theta0s_initial[1] - (alpha * overall_l1_dels0)
    
    thetas_final_l2 = thetas_initial[2] - (alpha * overall_l2_dels)
    
    theta0s_final_l2 = theta0s_initial[2] - (alpha * overall_l2_dels0)
    
    thetas_final_l3 = thetas_initial[3] - (alpha * overall_l3_dels)
    
    theta0s_final_l3 = theta0s_initial[3] - (alpha * overall_l3_dels0)
    
    
    
    
    neg_log_loss_value = neg_log_loss(C_batch,l3_activations)
    
    print("Pass # =",i,"Cross Entropy Loss =",neg_log_loss_value)
    
    iterations.append(i)
    
    neg_log_loss_history.append(neg_log_loss_value[0])
    
    thetas_initial[1] = thetas_final_l1
    
    theta0s_initial[1] = theta0s_final_l1
    
    thetas_initial[2] = thetas_final_l2
    
    theta0s_initial[2] = theta0s_final_l2
    
    thetas_initial[3] = thetas_final_l3
    
    theta0s_initial[3] = theta0s_final_l3
    
    i+= 1
```
